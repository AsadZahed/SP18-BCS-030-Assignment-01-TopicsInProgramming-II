# ***2) Main Driver code (Make sure to run all cells before calling Main Function)***

```python
#Np.arrange is used to initializing the values in evenly spaced manner.
#In this case i am initialing the value from -15 to 15(written 16 as python will exclude 16) and i have given interval as 1.
#np.arrange will return numpy array
x = np.arange(-15, 16, 1)
y = np.arange(-15, 16, 1)
print(x)
print(y)
```

### ***3) Task-1 Show a 30 * 30 Grid***

```python
def Task_1_Grid():
    #np.column_stack will take 1-D array of above initiazlied numpy array and then stack them into column to make it 2-D array of 30*30 grid points.
    xy_axis = np.column_stack([[i, j] for i in x for j in y])
    #Setting numpy printing threshold to maximum so that i can all the resultant 2D-array of np.column stack. 
    np.set_printoptions(threshold=sys.maxsize)
    #By using Matplotlib i am telling the size of grid to be 14 *10
    plt.figure(figsize=(14, 7))
    plt.title("Task-1: 30 * 30 Grid")
    plt.grid()
    #Scattering the points in 2D - Array into the grid
    plt.scatter(xy_axis[0],xy_axis[1])
    plt.show()
Task_1_Grid()
```

## ***4) Task-2 Shear tranform the matrix using loop my doing a dotproduct with each entry***

```python
def looped_shear_tranformation(x,y):
    #np.column_stack will take 1-D array of above initiazlied numpy array and then stack them into column to make it 2-D array of 30*30 grid points.
    xy_axis = np.column_stack([[i, j] for i in x for j in y])
    #By using for loop i am doing dot product with each point on the 2D array with 
    for k in range(xy_axis[0].size):
        xy_axis[0][k] += xy_axis[1][k]
    #By using Matplotlib i am telling the size of grid to be 14 * 7
    plt.figure(figsize=(14, 7))
    #Ploting the points in 2D - Array into the grid, making the color Dark Green using 'go'
    plt.plot(xy_axis[0], xy_axis[1],'go')
    plt.grid()
    plt.title("Task-2: Shear Tranformation using Dot product by Looping through Each Entry")
    plt.show()
looped_shear_tranformation(x,y)
```

### ***5)Task-3 Shear tranform the matrix parallely by using numpy***

```python
def numpy_parallel_shear_tranformation(x,y):
    xy_axis = np.column_stack([[i, j] for i in x for j in y])
    #By using a 2 * 2 matrix which contains shear tranformation generl matrix which has first column [1  0] and 2nd column [1  1]
    a=np.vstack([np.c_[1,1], np.c_[0,1]])
    #Now by using numpy dot product function, i am doing dotproduct of our xy_axis 2-d array with above intitialzied shear tranformation array.
    dot_product = np.dot(a, xy_axis)
    #By using Matplotlib i am telling the size of grid to be 14 * 7
    plt.figure(figsize=(14, 7))
    plt.grid()
    #Ploting the points in 2D - Array into the grid, making the color Dark Blue using 'bo'
    plt.plot(dot_product[0], dot_product[1],'bo')
    plt.title("Task-3: Shear Tranformation using Numpy Parallely")
    plt.show()
numpy_parallel_shear_tranformation(x,y)  
```
